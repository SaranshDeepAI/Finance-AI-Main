```python
import chromadb

client = chromadb.Client()
collection = client.create_collection("test_finance")

collection.add(
    documents=["Apple stock rose 3% after strong earnings report"],
    ids=["doc1"]
)

results = collection.query(
    query_texts=["What happened to Apple stock?"],
    n_results=1
)

print(results["documents"])
```
