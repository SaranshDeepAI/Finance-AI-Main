```python
newsapi = NewsApiClient(api_key="Replace With Yours")
print("NewsAPI connected successfully!")
```

```python
def get_stock_news(company_name, num_articles=5):
    news = newsapi.get_everything(
        q=company_name,
        language="en",
        sort_by="publishedAt",
        page_size=num_articles
    )
    
    articles = []
    for article in news["articles"]:
        articles.append({
            "title": article["title"],
            "description": article["description"],
            "source": article["source"]["name"],
            "published": article["publishedAt"],
            "url": article["url"]
        })
    
    return pd.DataFrame(articles)
```

```python
news_df = get_stock_news("Apple", num_articles=5)
print(news_df[["title", "source", "published"]])
```

```python
def get_stock_news(company_name, ticker, num_articles=5):
    query = f"{company_name} stock OR {ticker} earnings OR {company_name} finance"
    
    news = newsapi.get_everything(
        q=query,
        language="en",
        sort_by="publishedAt",
        page_size=num_articles
    )
    
    articles = []
    for article in news["articles"]:
        articles.append({
            "title": article["title"],
            "description": article["description"],
            "source": article["source"]["name"],
            "published": article["publishedAt"],
            "url": article["url"]
        })
    
    return pd.DataFrame(articles)

# Test it
news_df = get_stock_news("Apple", "AAPL", num_articles=5)
print(news_df[["title", "source", "published"]])
```
